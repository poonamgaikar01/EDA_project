```python
df['TIME']=pd.to_datetime(df['TIME'],format='%H:%M:%S').dt.time
```

```python
station_mask = (
                (df["STATION"] == "ATLANTIC AV") |
                (df["STATION"] == "FLATBUSH AV-B.C")|
                (df["STATION"] == "ATL AV-BARCLAY")|
                (df["STATION"] == "25 AV")|
                (df["STATION"] == "BERGEN ST")|
                (df["STATION"] == "FULTON ST")|
                (df["STATION"] == "LAFAYETTE AV")
               )
station_mask
```

```python
selected_stations=df[station_mask]
selected_stations
```

```python
selected_stations_data=selected_stations.drop(columns=['LINENAME','DIVISION','DESC'])
```

```python
selected_stations_data.head(200)
```

```python
Entries_byDate = selected_stations_data.groupby(['STATION','DATE']).ENTRIES.agg({'min_ENTRIES':'min'})
Entries_byDate = Entries_byDate.reset_index()
Entries_byDate['DAILY_ENTRIES'] = Entries_byDate.groupby(['STATION','DATE']).min_Entries.diff().shift(-1)
```

```python
Entries_byDate.loc[Entries_byDate['DAILY_ENTRIES'] <= 0, 'DAILY_ENTRIES'] = np.nan
Entries_byDate.loc[Entries_byDate['DAILY_ENTRIES'] > 100000, 'DAILY_ENTRIES'] = np.nan
Entries_byDate.head()
```
